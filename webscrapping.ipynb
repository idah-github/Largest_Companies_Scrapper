```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
page = requests.get(url)
pagecontent = BeautifulSoup(page.text, "html")
print(pagecontent.prettify())
```

```python
# get the first table
table1 = pagecontent.find_all('table')[0]
table2 = pagecontent.find('table',
                          class_="wikitable sortable sticky-header-multi sort-under")
print(table1.prettify())
```

```python
tblheads = table1.find_all('th')
tblheads
```

```python
# get the first relevant table heads
headings = [th.text.strip() for th in tblheads[1:9]]
print(headings)
```

```python
#add the th to a dataframe
df = pd.DataFrame(columns=headings)
df
```

```python
rows = table1.find_all('tr')
rows
```

```python
#get the td and populate the df
for row in rows[2:]:
    rowdata = row.find_all('td')
    rowdata1 = [data.text.strip() for data in rowdata]
#print(rowdata1)
    length = len(df)
    df.loc[length] = rowdata1
    
    
```
